```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\Goutam\OneDrive\Desktop\K_mean\Mall_Customers.csv")

df=df.drop(columns=['CustomerID',"Genre",'Age'])
df.info()
```

# Visual For unsuperWised

```python
sns.scatterplot(x='Annual Income (k$)',y='Spending Score (1-100)',data=df)
```

```python
from sklearn.cluster import KMeans
x=df.values
```

# Cheching Number of Cluster

```python
d_list=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=0)
    kmeans.fit(x)
    d_list.append(kmeans.inertia_)
print(d_list)
    
```

# Make Visual and take number of Cluster

```python
# jis place pr elbow bnana strt ho vhi number of Cluster rhta h
plt.plot(range(1,11),d_list)
plt.title("The Elbow Method")
plt.xlabel("Number  of cluster ")
plt.grid()
plt.ylabel("Distance")
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0)
kmeans.fit(x)
```

```python
x_pre=kmeans.predict(x)
```

#  Final Visual 

```python
plt.scatter(x[x_pre==0,0],x[x_pre==0,1],s=100,c='red',label='cluster_1')
plt.scatter(x[x_pre==1,0],x[x_pre==1,1],s=100,c='blue',label='cluster_2')
plt.scatter(x[x_pre==2,0],x[x_pre==2,1],s=100,c='magenta',label='cluster_3')
plt.scatter(x[x_pre==3,0],x[x_pre==3,1],s=100,c='green',label='cluster_4')
plt.scatter(x[x_pre==4,0],x[x_pre==4,1],s=100,c='blue',label='cluster_5')
```
